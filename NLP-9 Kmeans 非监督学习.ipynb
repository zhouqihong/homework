```python
%matplotlib inline
```

```python
import random
```

```python
x1 = [random.randint(-100, 100) for _ in range(100)]
x2 = [random.randint(-100, 100) for _ in range(100)]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x1, x2)
```

```python
trainning_data = [(x1,x2) for x1,x2 in zip(x1,x2)]
```

```python
from sklearn.cluster import KMeans
```

```python
cluster = KMeans(n_clusters=6)
```

```python
cluster.fit(trainning_data) # 下面的结果是什么意思？
```

```python
cluster.cluster_centers_   #?为什么要用两边用下划线呀？下面的结果是什么意思？
```

```python
cluster.labels_
```

```python
for center in cluster.cluster_centers_:
    plt.scatter(*center, s=100)  # s是什么？
plt.scatter(x1, x2, c='blue')  #c是什么？
```

```python
from collections import defaultdict
```

```python
centers = defaultdict(list)
```

```python
for label, location in zip(cluster.labels_, trainning_data):
    centers[label].append(location)
```

```python
color = ['red', 'green', 'grey', 'black', 'yellow', 'orange']

for i, c in enumerate(centers):
    for location in centers[c]:
        plt.scatter(*location, c=color[i])
        
for center in cluster.cluster_centers_:
    plt.scatter(*center, s=100)
```
